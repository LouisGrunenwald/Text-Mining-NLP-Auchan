```python
import pandas as pd
from projetpython import text_treatment as tt
```

```python
data = pd.read_csv("scrapped.csv", delimiter=';' , encoding='utf-8')
data.head()
```

```python
data.informations = data.informations.apply(lambda x : tt.filter_stropwords(tt.regex(x)) )
data.allergenes = data.allergenes.apply(lambda x : tt.regex(x))
data.score = data.score.apply(lambda x: x[-1] if type(x)==str else "na")
data["tags"] = data.allergenes.apply(lambda ligne : tt.tagger(ligne , tt.allergenes) )
data.head()
```

```python
treated = data.loc[ : , ['titres', 'urls', 'informations', 'allergenes', 'score',
       'tags']]
treated.to_csv("treated.csv", sep=";" , encoding='utf-8')
```
